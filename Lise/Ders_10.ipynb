```python
m_sayi = random.randint(1,100)
sayac =0
while True:
  if sayac >7:
    print("8 Hakkın da doldu")
    break
  sayac +=1
  sayi = int(input("1 ile 100 arasında bir sayı gir...(çıkmak için de 0 a bas)(tahmin hakkın 8 keredir)"))
  if sayi == 0:
    print("Oyun kapatılıyor..")
    time.sleep(3)
    print("Oyun kapatıldı...")
    break
  elif sayi < m_sayi:
    print("Daha yüksek bir sayı gir: ")
    continue
  elif sayi > m_sayi:
    print("Daha düşük bir sayı girmelisin: ")
    continue
    #sayi == m_sayi
  else:
    print("Seçtiğim sayı: ", sayi)
    print("Tahmin sayısı: ", sayac)
    break
```
